```python
import forgi
import forgi.visual.mplotlib as forgi_plot
import forgi.graph.bulge_graph as forgi_bg
import matplotlib.pyplot as plt

seq    = 'GCGGAUUUAGCUCAGUUGGGAGAGCGCCAGACUGAAGAUCUGGAGGUCCUGUGUUCGAUCCACAGAAUUCGCACCA'
struct = '(((((((....((((((.((......)).))))))....(((..((((........))))..))))))))))....'
fg = forgi_bg.BulgeGraph.from_dotbracket(seq=seq, dotbracket_str=struct)

fig,ax = plt.subplots(1, 1, figsize=(10,10))
forgi_plot.plot_rna(fg, ax=ax,
                    lighten=0.5,
                    text_kwargs={ "fontweight": "black" },
                    backbone_kwargs={ "linewidth": 1 })
plt.show()
```
